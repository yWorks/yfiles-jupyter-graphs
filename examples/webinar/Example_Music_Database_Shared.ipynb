## Display networkx-graphs



example1 (1) (1).png

**Step 1: Create a graph using networkx**

````python
import networkx as nx

def createCollaborationGraph():
  G = nx.Graph()
    # Add artist nodes
  for artist in data["artists"]:
      G.add_node(
          artist['id'],
          label=artist['name'],
          type='artist',
          data=artist
      )

  # Add group nodes
  for group in data["groups"]:
      G.add_node(
          group['id'],
          label=group['name'],
          type='group',
          data=group
      )

  # Add edges between artists and their groups
  for artist in data["artists"]:
      for group_id in artist["groups"]:
          if G.has_node(group_id):
              G.add_edge(artist['id'], group_id)

  return G
````

Step 2: Visualize the graph using yFiles

````python
collaborationGraph = createCollaborationGraph()

w = GraphWidget(graph=collaborationGraph)

w.set_edge_styles_mapping(lambda edge: {"color": "#D90368"})
w.set_node_styles_mapping(lambda node: {"color": "#a367dc"} if node["properties"]["type"] == "group" else {"color": "#67b7dc","shape": "rectangle"})


w.show()
````

## A more sophisticated example with networkx

Show all collaborations of a single artist, e.g., Paul McCartney

**Step 1: Function that returns all collaborations of a given artist.**

````python
def getCollaborations(G, artist_name, max_depth):
    # Find the starting artist node by name
    start_node = next((n for n, d in G.nodes(data=True) if d.get('label') == artist_name), None)

    reachable_nodes = nx.single_source_shortest_path_length(G, start_node, cutoff=max_depth)
    nodes = [node for node in G.nodes if node in reachable_nodes]
    # Return subgraph induced by nodes within 2 hops
    return G.subgraph(nodes).copy()
````

**Step 2: Show the networkx-graph with yFiles**

````python
subgraph = getCollaborations(collaborationGraph,"Paul McCartney",2)

w = GraphWidget(graph=subgraph)

w.set_edge_styles_mapping(lambda edge: {"color": "#D90368"})
w.set_node_styles_mapping(lambda node: {"color": "#a367dc"} if node["properties"]["type"] == "group" else {"color": "#67b7dc","shape": "rectangle"})
w.radial_layout()

w.show()
````

## Spatial relations

**Step 1: Function that returns the collaboration graph of a given artist (without music groups).**

````python
from itertools import combinations
from collections import defaultdict
def build_artist_collaboration_graph():
    # Create a mapping from group ID to list of artist IDs
    group_to_artists = defaultdict(set)
    for artist in data["artists"]:
        for group_id in artist['groups']:
            group_to_artists[group_id].add(artist['id'])

    # Create a graph for collaborations
    G = nx.Graph()

    # Add artist nodes with metadata
    for artist in data["artists"]:
        G.add_node(artist['id'], label=artist['name'], type='artist', data = artist)

    # Add edges between artists who share a group
    for group_artists in group_to_artists.values():
        for artist1, artist2 in combinations(group_artists, 2):
            if not G.has_edge(artist1, artist2):
                G.add_edge(artist1, artist2, weight=1)

    return G
````

````python
simpleCollaborationGraph = build_artist_collaboration_graph()

w = GraphWidget(graph=simpleCollaborationGraph)


w.set_node_coordinate_mapping(lambda node: node["properties"]["data"]["coordinates"])
w.set_node_styles_mapping(lambda node: {"color": "#a367dc"})
w.set_node_size_mapping(lambda node: (10,10))

w.map_layout()
w.show()
````

## Spatial Relations with Heat Map

Annotate nodes with heat, i.e., the relative number of releases of an artist

````python
simpleCollaborationGraph = build_artist_collaboration_graph()

w = GraphWidget(graph=simpleCollaborationGraph)


w.set_node_coordinate_mapping(lambda node: node["properties"]["data"]["coordinates"])
w.set_node_styles_mapping(lambda node: {"color": "#a367dc"})
w.set_node_size_mapping(lambda node: (10,10))

maxReleases = max([artist["releases"] for artist in data["artists"]])

w.set_heat_mapping(lambda element: element["properties"]["data"]["releases"]/maxReleases if "data" in element["properties"]  else 0)

w.map_layout()
w.show()
````
